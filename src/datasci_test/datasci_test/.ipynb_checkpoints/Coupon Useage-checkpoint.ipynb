```python
from dbconnector import *
import pandas as pd
import numpy as np
%matplotlib inline
%gui
```

```python
dbconfig = get_configuration()
connection = make_connection(dbconfig)
q_devices = "Select * from datasci.devices"
q_actions = "Select * from datasci.actions"
q_products_lookup = "Select * from datasci.products_lookup"

devices = query_db(connection, q_devices)
actions = query_db(connection, q_actions)
product_lookup = query_db(connection, q_products_lookup)
```

```python
q_coupons = "SELECT t1.*, t2.item_name \
FROM datasci.devices as t1 \
Inner Join datasci.products_lookup as t2 on t2.item_id = t1.item_id \
where length(t1.coupon_code)>0;"

#another option for q_coupon
"""SELECT t1.*,
t2.coupon_code,
t2.time_stamp as registration_timestamp,
t2.item_registration_id as reg_id_2
FROM datasci.actions as t1
inner join datasci.devices as t2 on t1.user_id= t2.user_id
where length(t1.item_registration_id)>0 and t1.item_name like 'Status %'
"""
coupons = query_db(connection,q_coupons)
```

```python
coupons2 = pd.merge(devices,product_lookup,on='item_id')
```

```python
coupons2 = coupons2[coupons2['coupon_code'].map(lambda x: len(x)>0)]
```

```python
pd.pivot_table(coupons2,index='time_stamp',aggfunc='count',values = 'coupon_code').plot()
```
